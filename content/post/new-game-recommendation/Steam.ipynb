```python
datashape=(len(playtime['The Elder Scrolls V Skyrim'].unique()),len(playtime['151603712'].unique()))
sparsedata=pd.DataFrame(np.zeros(datashape),index=playtime['The Elder Scrolls V Skyrim'].unique(),columns=playtime['151603712'].unique())
for i in range(len(playtime)):
    item=playtime.iloc[i]
    sparsedata[item['151603712']][item['The Elder Scrolls V Skyrim']]=item['1.0']
    #スパースなデータ行列完成
    #本当はもっといいやり方があるだろうが、そんなに実行時間はかからないしこれで済ませる
sparsedata
```

```python
Me=np.zeros(len(playtime['The Elder Scrolls V Skyrim'].unique()))
Me[38]=50.0 #Undertale 50 hours
Me[0]=80.0 #TESV 80 hours
```

```python
item_resembrance=np.dot(sparsedata,sparsedata.T)
item_resembrance=pd.DataFrame(item_resembrance,index=playtime['The Elder Scrolls V Skyrim'].unique(),columns=playtime['The Elder Scrolls V Skyrim'].unique())
```

```python
item_resembrance['The Elder Scrolls V Skyrim'].sort_values(ascending=False)
```

```python
item_resembrance['Undertale'].sort_values(ascending=False)
#あれ？undertaleより他のゲームのほうが人気？
#単純にプレイ時間が長いことにより内積が大きくなっていた模様
```

```python
normalized_sparsedata=np.array(sparsedata)/np.expand_dims(np.max(sparsedata,axis=1),axis=1)
#最大値で割ることでnormalizeする
```

```python
normalized_sparsedata
```

```python
normalized_item_resembrance=np.dot(normalized_sparsedata,normalized_sparsedata.T)
normalized_item_resembrance=pd.DataFrame(normalized_item_resembrance,index=playtime['The Elder Scrolls V Skyrim'].unique(),columns=playtime['The Elder Scrolls V Skyrim'].unique())
```

```python
normalized_item_resembrance['The Elder Scrolls V Skyrim'].sort_values(ascending=False)
```

```python
normalized_item_resembrance['Undertale'].sort_values(ascending=False)
```
